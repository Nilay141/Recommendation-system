```python
df['Updated_Text']=df["Text"]
df
```

```python
for i in range(len(df)):
    temp=df[df.index==i]['Text'].values[0]
    token=word_tokenize(temp)
    remove_stopwords=[i for i in token if not i in stopwords.words()]
    df[df.index==i]['Updated_Text'].values[0]=(" ").join(remove_stopwords)
```

```python
def sentiment(s):
    try:
        return TextBlob(str(s)).sentiment.polarity
    except:
        return 0
```

```python
df['Sentiment']=df['Updated_Text'].apply(sentiment)
```

```python
df=df.dropna()
df
```
